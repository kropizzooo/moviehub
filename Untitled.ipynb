```python
missing_values_count = bommoviegross.isnull().sum()
print("\nMissing values count:")
print(missing_values_count)


for col in bommoviegross.columns:
    unique_values = bommoviegross[col].unique()
    print(f"\nUnique values in '{col}':")
    print(unique_values)

    
    if col in ["worldwide_gross", "domestic_gross", "foreign_gross"]:
        invalid_values = bommoviegross[bommoviegross[col] < 0][col]
        if len(invalid_values) > 0:
            print(f"\nInvalid (negative) values in '{col}':")
            print(invalid_values)
```

```python
bommoviegross['foreign_gross'] = bommoviegross['foreign_gross'].astype(int)
bommoviegross['domestic_gross'] = bommoviegross['domestic_gross'].astype(int)


mean_foreign_gross = bommoviegross["foreign_gross"].mean()
median_foreign_gross = bommoviegross["foreign_gross"].median()

mean_domestic_gross = bommoviegross["domestic_gross"].mean()
median_domestic_gross = bommoviegross["domestic_gross"].median()


print(f"\nMean of foreign_gross: {mean_foreign_gross}")
print(f"Median of foreign_gross: {median_foreign_gross}")
print(f"\nMean of domestic_gross: {mean_domestic_gross}")
print(f"Median of domestic_gross: {median_domestic_gross}")
```

```python
iqr_foreign_gross =  bommoviegross["foreign_gross"].quantile(0.75) -  bommoviegross["foreign_gross"].quantile(0.25)
iqr_domestic_gross =  bommoviegross["domestic_gross"].quantile(0.75) -  bommoviegross["domestic_gross"].quantile(0.25)
threshold_foreign_gross = 1.5 * iqr_foreign_gross
threshold_domestic_gross = 1.5 * iqr_domestic_gross

outliers_foreign_gross = []
for i, row in  bommoviegross.iterrows():
    if abs(row["foreign_gross"] - mean_foreign_gross) > threshold_foreign_gross:
        outliers_foreign_gross.append(i)

outliers_domestic_gross = []
for i, row in  bommoviegross.iterrows():
    if abs(row["domestic_gross"] - mean_domestic_gross) > threshold_domestic_gross:
        outliers_domestic_gross.append(i)
print(f"\nOutliers in foreign_gross: {outliers_foreign_gross}")
print(f"\nOutliers in domestic_gross: {outliers_domestic_gross}")
```

```python
bommoviegross["total_gross"] = bommoviegross["foreign_gross"] + bommoviegross["domestic_gross"]

# Print the new DataFrame
print(bommoviegross)
```
